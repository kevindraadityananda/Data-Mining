```python
#Import library yang diperlukan
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Import dataset
data = pd.read_excel("Online_Retail.xlsx")
data.head()
```

```python
#Eksplor data kolom
data.columns
```

```python
#Eksplor wilayah transaksi
data.Country.unique()
```

```python
#Menghapus ruang ekstra dalam deskripsi
data['Description'] = data['Description'].str.strip()

#Menghapus baris tanpa nomor faktur
data.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
data['InvoiceNo'] = data['InvoiceNo'].astype('str')

#menghapus transaksi yang dilakukan secara kredit
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
#Memisahkan data sesuai wilayah transaksi
#Prancis
basket_France = (data[data['Country'] == "France"]
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))

#UK
basket_UK = (data[data['Country'] == "United Kingdom"]
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))

#Portugal
basket_Porugal = (data[data['Country'] == "Portugal"]
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))


#Sweden
basket_Sweden = (data[data['Country'] == "Sweden"]
                 .groupby(['InvoiceNo', 'Description'])['Quantity']
                 .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))

#Cek Hasil
basket_UK
```

```python
#Pengkodean data dengan "Hot Enkoding"
#Mendefinisikan fungsi "Hot Encoding" 
def hot_encode(x):
  if(x<=0):
    return 0
  if(x>=1):
    return 1
```

```python
#mengkode dataset
basket_encode = basket_France.applymap(hot_encode)
basket_France = basket_encode

basket_encode = basket_UK.applymap(hot_encode)
basket_UK = basket_encode

basket_encode = basket_Porugal.applymap(hot_encode)
basket_Porugal = basket_encode

basket_encode = basket_Sweden.applymap(hot_encode)
basket_Sweden = basket_encode

basket_Sweden
```

```python
#membangun model dan analisis hasil

#France
frq_items = apriori(basket_France, min_support=0.05, use_colnames=True)

rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```

```python
#UK
frq_items = apriori(basket_UK, min_support=0.01, use_colnames=True)

rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head(10)
```

```python
#Portugal
frq_items = apriori(basket_Porugal, min_support=0.05, use_colnames=True)

rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```

```python
#Sweden
frq_items = apriori(basket_Sweden, min_support=0.05, use_colnames=True)

rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```
