```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('Dataset_AHC.xlsx')
print(data)
```

```python
x = np.asarray(data)
print(x)
```

```python
#Complete Linkage
z = linkage(x, method='complete')
```

```python
print(z)
```

```python
plt.figure(figsize=(15,8))
plt.title('Dendrogram Hierachical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(z)
plt.show()
```

```python
#Average Linkage
z= linkage(x, method='average')
print(z)
```

```python
plt.figure(figsize=(15,8))
plt.title('Dendrogram Hierachical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(z)
plt.show()
```
